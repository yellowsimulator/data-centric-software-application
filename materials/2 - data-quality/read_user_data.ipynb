# Import librairies

```python
import json
import pandas as pd
from glob import glob
from pandas import json_normalize
```

```python
path = '../log_data/*/*/*.json'
user_files = glob(path)
user_files
```

# Inspection of one file

```python
user_file = user_files[0]
user_file
```

```python
with open(user_file, 'r') as file_object_list:
    for k, record in enumerate(file_object_list):
        data = json.loads(record)
        print(' ')
        print(f'{k}, {data}')
```

# Generating a dataframe for all files

```python
container = []
for user_file in user_files:
    with open(user_file, 'r') as file_object_iterable:
        for record in file_object_iterable:
            data = json.loads(record)
            df_temp = pd.json_normalize(data)
            container.append(df_temp)
df = pd.concat(container, ignore_index=True)
df
```
